```python
G ={
    'A':['B', 'C'],
    'B':['A', 'D', 'E'],
    'C':['A', 'F'],
    'D':['B','G'],
    'E':['B', 'H'],
    'F':['C'],
    'G':['D'],
    'H':['E'],
    
    
}
```

```python
def DFS(G,start):
    stack = []
    visited = []
    stack.append(start)
    while (stack):
        poppedVertex = stack.pop()
        visited.append(poppedVertex)
        for chimeki in G[poppedVertex]:
            if chimeki not in visited and chimeki not in stack:
                stack.append(chimeki)
                
    return visited
    
```

```python
start = 'A'
traverse_sequence = DFS(G,start)
print("DFS traversal of the graph is: ", traverse_sequence)
```
