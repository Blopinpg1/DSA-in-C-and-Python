```python
G = {
    'V1': {'V2': 1, 'V4': 4},
    'V2': {'V1': 1, 'V3': 3, 'V5': 3},
    'V3': {'V2': 3, 'V5': 1, 'V6': 2},
    'V4': {'V1': 4, 'V5': 1},
    'V5': {'V4': 1, 'V2': 3, 'V3': 1, 'V6': 4},
    'V6': {'V3': 2, 'V5': 4}
}
```

```python
from queue import PriorityQueue

def primsAlgorithm(G, start):
    visited = set() 
    mst = []     
    total_cost = 0

    PQ = PriorityQueue()

    # Add all edges from the start vertex
    visited.add(start)
    for neighbor, weight in G[start].items():
        # print(f"Adding edge ({start}, {neighbor}) with weight {weight} to the priority queue.")
        PQ.put((weight, start, neighbor)) 
        # print(f"Visited vertices: {visited}") 

    while not PQ.empty() and len(visited) < len(G):
        # print("Current priority queue:", list(PQ.queue))
        weight, u, v = PQ.get()  
        if v not in visited:
            visited.add(v) 
            # print(f"Adding edge ({u}, {v}) with weight {weight} to the MST.")
            mst.append((u, v, weight)) 
            total_cost += weight  
            # Add all edges from the newly visited vertex
            for neighbor, wt in G[v].items():
                # print(f"Checking edge ({v}, {neighbor}) with weight {wt} .")
                if neighbor not in visited:
                    PQ.put((wt, v, neighbor))
                    # print(f"Adding edge ({v}, {neighbor}) with weight {wt} to the priority queue.")

    return mst, total_cost
```

```python
mst, cost = primsAlgorithm(G, 'V1')

print("Edges in MST:")
for u, v, w in mst:
    print(f"{u} - {v} with weight {w}")

print("\nTotal Cost:", cost)
```
