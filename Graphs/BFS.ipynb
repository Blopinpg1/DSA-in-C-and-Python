```python
G ={
    'A':['B', 'C'],
    'B':['A', 'D', 'E'],
    'C':['A', 'F'],
    'D':['B','G'],
    'E':['B', 'H'],
    'F':['C'],
    'G':['D'],
    'H':['E'],
    
    
}
```

```python
def BFS(G,start):
    queue = []
    visited = []
    queue.append(start)
    while (queue):
        outVertex = queue.pop(0)
        visited.append(outVertex)
        for chimeki in G[outVertex]:
            if chimeki not in visited and chimeki not in queue:
                queue.append(chimeki)
                
    return visited
    
```

```python
start = 'A'
traverse_sequence = BFS(G,start)
print("BFS traversal of the graph is: ", traverse_sequence)
```
