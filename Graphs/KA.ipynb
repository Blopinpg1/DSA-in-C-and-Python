```python
G = {
    'V1': {'V2': 1, 'V4': 4},
    'V2': {'V1': 1, 'V3': 3, 'V5': 3},
    'V3': {'V2': 3, 'V5': 1, 'V6': 2},
    'V4': {'V1': 4, 'V5': 1},
    'V5': {'V4': 1, 'V2': 3, 'V3': 1, 'V6': 4},
    'V6': {'V3': 2, 'V5': 4}
}
```

```python
def find(parent, node):
    # Find the root of the set where node belongs, with path compression
    if parent[node] != node:
        parent[node] = find(parent, parent[node])
    return parent[node]
```

```python
def union(parent, u, v):
    # Union two sets by updating the root of v to root of u
    root_u = find(parent, u)
    root_v = find(parent, v)
    if root_u != root_v:
        parent[root_v] = root_u
        return True
    return False
```

```python
def KA(graph):
    # Step 1: Create a list of all edges
    edges = []
    for u in graph:
        for v, w in graph[u].items():
            if (v, u, w) not in edges:  # avoid duplicates
                edges.append((u, v, w))

    # Step 2: Sort edges by weight
    edges.sort(key=lambda x: x[2])

    # Step 3: Create a parent dictionary for disjoint sets
    parent = {v: v for v in graph}
    print(parent)

    mst = []
    total_cost = 0

    # Step 4: Process edges in increasing order of weight
    for u, v, w in edges:
        if union(parent, u, v):
            mst.append((u, v, w))
            total_cost += w

    return mst, total_cost
```

```python
mst, cost = KA(G)
print("Edges in MST:")
for u, v, w in mst:
    print(f"{u} - {v} with weight {w}")
print("\nTotal Cost:", cost)
```
