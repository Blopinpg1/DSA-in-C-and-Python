```python
import math
G = [
    [0, 10, math.inf, 5, math.inf],
    [math.inf, 0, 1, 2, math.inf],
    [math.inf, math.inf, 0, math.inf, math.inf],
    [math.inf, math.inf, 9, 0, 2],
    [7, math.inf, 6, math.inf, 0]
]
```

```python
def FW(W):
    n = len(W)
    D = dict()
    D[0] = W
    for k in range(n):
        D[k+1] = D[k]
        for i in range(n):
            for j in range(n):
                D[k+1][i][j] = min(
                    D[k][i][j] , D[k][i][k] + D[k][k][j]
                )
                
    return (D[k+1])

        
```

```python
def main():
    W = G
    D = FW(W)
    print("The distance matrix is:")
    for row in D:
        print(row)
        
if __name__ == "__main__":
    main()

    
```
